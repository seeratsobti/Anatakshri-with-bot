```python
import random #to generate random string 
import  string #to do operation on strings
import time #to get time
import pandas as pd #to store data in pandas datatframe
import numpy as np #apply operations on data
import speech_recognition as sr     #record the speech
import pafy # perform search
import vlc #open vlc instance
from googlesearch import search #to do googlesearch
from datetime import datetime, timedelta #get the date and time
def random_letter():
    import random #to generate random string 
    import  string #to do operation on strings
    start_letter = random.choice(string.ascii_letters) #generate random letter
    letter=start_letter.lower() #convert it to lower case
    print('sing a song with the alphabet:') #print statement
    print(letter) #print letter generated
    return(letter)
def get_database():
    import pandas as pd #to store data in pandas datatframe
    import numpy as np #apply operations on data
    raw_data =  pd.read_csv("F:/NOTES/aiml lab/billboard_lyrics_1964-2015.csv",encoding='cp1252') #import data
    raw_data=raw_data.drop(['Year','Artist','Source','Rank'],axis=1) #drop extra columns
    raw_data.Lyrics = raw_data.Lyrics.str.lstrip()  #space in starting of lyrics
    raw_data.Lyrics = raw_data.Lyrics.str.rstrip() #space in last of lyrics
    raw_data['first_letter'] = raw_data['Lyrics'].astype(str).str[0] #get first letter of lyrics
    raw_data['new_letter'] = raw_data['Lyrics'].astype(str).str[-1] #get last letter of lyrics
    raw_data = raw_data.dropna(axis = 0, how ='any') #drop if any null value
    return(raw_data)
def recorder(): 
    import speech_recognition as sr     #record the speech
    r = sr.Recognizer()                 # initialize recognizer
    with sr.Microphone() as source:     # mention source as Microphone
        print("Speak Anything :")
        audio = r.listen(source)        # listen to the source
        try:
            text = r.recognize_google(audio)    # use recognizer to convert our audio into text part.
            text=text.lower()     #convert text to lowercase
            print("You said : {}".format(text)) #print
        except:
            print("Sorry could not recognize your voice") #print
            return recorder() #playn untill speech is recognised
            
    return(text)
def playsong(song):
    import time #to get time
    import pafy # perform search
    import vlc #open vlc instance
    from googlesearch import search #to do googlesearch
    from datetime import datetime, timedelta #get the date and time
    query=str(list(song['Song']))+(' song') #generate the staement to be used as query
    for url in search(query,tld='com',lang='en',num=1,stop=1,pause=2): #search url
        video= pafy.new(url) #gets url
        best=video.getbest() #select best video url
        playurl= best.url #save url
        videotime=video.length #get length of video
        #end_time= datetime.now()+timedelta(seconds=10)
        #while datetime.now()< end_time:
        Instance=vlc.Instance() #open vlc instance
        player=Instance.media_player_new() 
        Media=Instance.media_new(playurl) #play url
        Media.get_mrl()
        player.set_media(Media)
        player.play() #play song
        time.sleep(videotime) #sleep instance after video time ends
        player.stop() #close vlc
    return(player)
def check(player_score,bot_score,letter):
    letter=letter
    raw_data=get_database()
    
    text=recorder()
    check_letter_one=text[0][:1] #check letter matches or not
    if check_letter_one== letter:
        player_score=player_score+1
        second_letter=text[-1][:1] #get second start letter
        subset=raw_data[raw_data['first_letter']==second_letter] #select all songs with second start letter
        
        song=subset.sample(n=1,replace=True) #select one song from sample
        letter=song['new_letter']
        playsong(song)   
        #print(song)
        bot_score=bot_score+1
        print('sing song with letter',letter)
    else:
        print('please sing a song with letter')#print statement
        print(letter)
        return(check(player_score,bot_score,letter))
        #check(letter)
    return(player_score,bot_score,letter)
def main():
    player_score=0 #initialise score
    bot_score=0 #initialise score
    letter=random_letter() #get random letter
    while (player_score<5 or bot_score<5): #winning condition
        player_score,bot_score,letter=check(player_score,bot_score,letter) #run functionn multiple times
    if(player_score>bot_score): #decide winner
        print("Congratulations!you won!") #winning message
    else:
        print('Sorry! you lost!')#loosing message
if __name__=='__main__':
    main()
```
